# Exploration du fichier CSV des manifestations

L'analyse concerne les manifestations en Suisse
<br/>

NB le fichier n'est pas disponible sur le serveur publiquement, il doit être déposé dans le dossier 'comoplements_FB' et il est exclu du versionnement GIT 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_add = 'poldem-protest_6_utf8.csv'
df = pd.read_csv(file_add, sep='\t', header=0)
print(len(df))
df
```

```python
df_ch = df[df.country == 'CH'] 
print(len(df_ch))
df_ch.head()
```

```python
df_ch.info()
```

```python
df_fr = df[df.country == 'F'] 
print(len(df_fr))
df_fr.head()
```

```python
ax = df_ch.groupby(by='year', observed=True).size()\
      .plot(kind='bar',rot=60, fontsize=8, figsize=(12,8),color='lightblue')

ax.bar_label(ax.containers[0], fontsize=9)

plt.ylabel('Effectif')
plt.xlabel('Année')
plt.title('Effectifs annuels des manifestations Suisse')
plt.show()
```

```python
ax = df_ch.groupby(by='form', observed=True).size().sort_values(ascending=True)\
      .plot(kind='barh',rot=10, fontsize=8, figsize=(8,12),color='lightblue')

ax.bar_label(ax.containers[0], fontsize=9)

plt.ylabel('Type')
plt.xlabel('Effectif')
plt.title('Distribution par type de manifestation')
plt.show()
```
